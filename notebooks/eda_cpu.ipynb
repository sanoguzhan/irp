```python
import os
import pandas as pd
from pathlib import Path
import seaborn as sns
import plotly.express as px
```

```python
sns.set_style("ticks")

sns.set_context("paper")
base_dir = Path().cwd().parent
```

```python
### File paths
container_data = base_dir / "data" / "container"
vm_data = base_dir / "data" / "vm"
ec2_data = base_dir / "data" / "ec2"

def get_files(root_dir:Path, parent_folder:str, folder_name: str) -> Path:
    path = root_dir / parent_folder / folder_name / "stats" / "csv"
    return {x.stem:x for x in path.iterdir()}

container_nginx_cpu_data = get_files(container_data, "nginx", "cpu")
container_apache_cpu_data = get_files(container_data, "apache", "cpu")


vm_nginx_cpu_data = get_files(vm_data, "nginx", "cpu")
vm_apache_cpu_data = get_files(vm_data, "apache", "cpu")


ec2_nginx_cpu_data = get_files(ec2_data, "nginx", "cpu")
ec2_apache_cpu_data = get_files(ec2_data, "apache", "cpu")
```

```python
container_nginx_cpu_periods ={'Period 1': 100 - container_df_nginx_cpu_1["id"] ,
    'Period 2': 100 - container_df_nginx_cpu_2["id"] ,
    'Period 3': 100 - container_df_nginx_cpu_3["id"] }
container_nginx_cpu_total_average = pd.DataFrame(container_nginx_cpu_periods).mean(axis=1)
container_nginx_cpu_total_average.plot()
```

```python
n=1
num_rows = 115
seconds = list(range(5, 5 + num_rows, 5))

def transform_data(df, type_name):
    df["Type"] = type_name
    df = df[["Type", "id"]].rename(columns={"id": "CPU Used (%)"})
    df["CPU Used (%)"] = 100 - df["CPU Used (%)"] 
    df["Seconds"] = seconds 
    return df

container_nginx_cpu_idle = pd.read_csv(container_nginx_cpu_data["vmstats-2C0"], skiprows=1)
container_nginx_cpu_idle.drop(container_nginx_cpu_idle.tail(n).index,inplace=True)
container_nginx_cpu_idle.drop(container_nginx_cpu_idle.head(n).index,inplace=True)
container_nginx_cpu_idle = transform_data(container_nginx_cpu_idle, "Container - Nginx")



container_apache_cpu_idle = pd.read_csv(container_apache_cpu_data["vmstats-2C0"], skiprows=1)
container_apache_cpu_idle.drop(container_apache_cpu_idle.tail(n).index,inplace=True)
container_apache_cpu_idle.drop(container_apache_cpu_idle.head(n).index,inplace=True)
container_apache_cpu_idle = transform_data(container_apache_cpu_idle, "Container - Apache")


vm_nginx_cpu_idle= pd.read_csv(vm_nginx_cpu_data["vmstats-2C0"], skiprows=1)
vm_nginx_cpu_idle.drop(vm_nginx_cpu_idle.tail(n).index,inplace=True)
vm_nginx_cpu_idle.drop(vm_nginx_cpu_idle.head(n).index,inplace=True)
vm_nginx_cpu_idle = transform_data(vm_nginx_cpu_idle, "Virtual Machine - Nginx")


vm_apache_cpu_idle = pd.read_csv(vm_apache_cpu_data["vmstats-2C0"], skiprows=1)
vm_apache_cpu_idle.drop(vm_apache_cpu_idle.tail(2).index,inplace=True)
vm_apache_cpu_idle.drop(vm_apache_cpu_idle.head(2).index,inplace=True)
vm_apache_cpu_idle = transform_data(vm_apache_cpu_idle, "Virtual Machine - Apache")


ec2_nginx_cpu_idle = pd.read_csv(ec2_nginx_cpu_data["vmstats-2C0"], skiprows=1)
ec2_nginx_cpu_idle.drop(ec2_nginx_cpu_idle.tail(3).index,inplace=True)
ec2_nginx_cpu_idle.drop(ec2_nginx_cpu_idle.head(2).index,inplace=True)
ec2_nginx_cpu_idle = transform_data(ec2_nginx_cpu_idle, "Cloud Instance - Nginx")


ec2_apache_cpu_idle= pd.read_csv(ec2_apache_cpu_data["vmstats-2C0"], skiprows=1)
ec2_apache_cpu_idle.drop(ec2_apache_cpu_idle.tail(3).index,inplace=True)
ec2_apache_cpu_idle.drop(ec2_apache_cpu_idle.head(2).index,inplace=True)
ec2_apache_cpu_idle = transform_data(ec2_apache_cpu_idle, "Cloud Instance - Apache")



df_all_cpu_idle = pd.concat([container_nginx_cpu_idle,
                                 container_apache_cpu_idle,
                                 vm_nginx_cpu_idle,
                                 vm_apache_cpu_idle,
                                 ec2_nginx_cpu_idle,
                                 ec2_apache_cpu_idle])

fig = px.line(df_all_cpu_idle, x="Seconds", y="CPU Used (%)", color="Type")

fig.show()
```
